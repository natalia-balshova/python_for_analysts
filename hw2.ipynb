#### 3.1 В каком диапазоне изменяются стоимости недвижимости?   

```python
min_price = df['price'].min() # минимальная стоимость дома
min_price
```

```python
max_price = df['price'].max() # максимальная стоимость дома
max_price
```

```python
median_price = df['price'].quantile(0.5) # медиана (средняя стоимость дома)
median_price
```

```python
mean_price = df['price'].mean() # средняя стоимость дома
mean_price
```

```python
print(f'Стоимости недвижимости изменяются в диапазоне от {min_price} до {max_price}, средняя стоимость {mean_price.round(2)}.')
```
